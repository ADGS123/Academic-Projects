2) Calcula e imprime la media y la desviación estándar para cada una de las 4 columnas de medición (es decir, la longitud y el ancho del sépalo, la longitud y el ancho del pétalo). Recuerda que, para una lista dada den valores x = (x1, x2, ..., xn), la media μ y la desviación estándar σ se definen respectivamente como:

```python

#Cols solo recorre desde 0 hasta 3
import math as mt

#Col[0]= longitud del sépalo (cm)
#Col[1]=Ancho del sépalo (cm)
#Col[2]=longitud del pétalo
#Col[3]=Ancho del pétalo

DAtos=['longitud del sépalo','ancho del sépalo','longitud del pétalo','ancho del pétalo'] #Lista de características

prom=[] #lista vacía para guardar los valores promedios
desvest=[] #lista vacía para guardar los valores de desviación estándar

#ciclo para valores promedio
for j in range(4): #Repetimos cambiando de columna
    s=0 #contador para la suma
    with open(r"C:\Users\DAVID\Desktop\iris\iris.data")  as f: #abrimos el archivo de texto
            for cols in csv.reader(f):#recorremos los datos por filas
                s=s+float(cols[j]) #sumamos cada dato de cada columna respectivamente 
            prom.append(s/150) #pegamos el valor promedio en la lista
#ciclo para valores de desviación estándar
for j in range(4): #Repetimos por el número de columnas
    s1=0 #contador para la suma interna de desviación estándar
    with open(r"C:\Users\DAVID\Desktop\iris\iris.data")  as f: #volvemos a abrir el archivo de texto
        for cols in csv.reader(f): #volvemos a leerlo por filas
            s1=s1+(float(cols[j])-prom[j])**2 #hacemos talacha de la desviación estándar
        desvest.append(mt.sqrt(s1/150)) #pegamos los resultados en su lista
    print('Valor promedio de',DAtos[j],'=',prom[j],'cm') # establecemos un formado para que imprima los valores
print("\n") #brinco para que no se vean muy amontonados los valores

for j in range(4): #ciclo para imprimir las desviaciones estándar
    print('Desviación estándar de',DAtos[j],'=',desvest[j],'cm')
```

3) Calcula e imprime la media y la desviación estándar para cada una de las 4 columnas de medición, por separado para cada una de las tres especies de Iris (versicolor, virginica y setosa).

```python

#Col[0]= longitud del sépalo (cm)
#Col[1]=Ancho del sépalo (cm)
#Col[2]=longitud del pétalo
#Col[3]=Ancho del pétalo

#listas para guardar los valores promedio de las 4 columnas de las respectivas flores
setosa=[]
versicolor=[]
virginica=[]

#listas para guardar los valores de desviación estándar de las 4 columnas de las respectivas flores
dessetosa=[]
desversi=[]
desvirgi=[]


#ciclo for para recorrer las columnas
for j in range(4): #Repetimos cambiando de columna
    s1=0
    s2=0
    s3=0  #contadores para las sumas de cada tipo de flor
    with open(r"C:\Users\DAVID\Desktop\iris\iris.data")  as f: #abrimos el archivo de texto
            for cols in csv.reader(f):#recorremos los datos por filas
                if cols[4]=="Iris-setosa": #condición de la primer flor
                    s1=s1+float(cols[j]) #sumamos cada dato de cada columna respectivamente 
                elif cols[4]=='Iris-versicolor': #condición para la segunda flor
                     s2=s2+float(cols[j]) #sumamos otra vez
                elif cols[4]=='Iris-virginica': #condición para la tercera flor
                    s3=s3+float(cols[j]) #sumamos para la última flor
            setosa.append(s1/50) #pegamos los valores promedio en las listas
            versicolor.append(s2/50)
            virginica.append(s3/50)
                



#código para la desviación estándar
for j in range(4): #Repetimos por el número de columnas
    s1=0
    s2=0
    s3=0 #contadores para la suma interna de desviación estándar
    with open(r"C:\Users\DAVID\Desktop\iris\iris.data")  as f: #volvemos a abrir el archivo de texto
        for cols in csv.reader(f): #volvemos a leerlo por filas
            if cols[4]=="Iris-setosa": #condición de la primer flor
                 s1=s1+(float(cols[j])-setosa[j])**2 #sumamos cada dato de cada columna respectivamente 
            elif cols[4]=='Iris-versicolor': #condición para la segunda flor
                 s2=s2+(float(cols[j])-versicolor[j])**2
            elif cols[4]=='Iris-virginica': #condición para la tercera flor
                 s3=s3+(float(cols[j])-virginica[j])**2
        dessetosa.append(mt.sqrt(s1/50)) #pegamos los resultados en su lista
        desversi.append(mt.sqrt(s2/50))
        desvirgi.append(mt.sqrt(s3/50))

    
List_prom=[setosa,versicolor,virginica] #Lista de valores promedio de las 4 características por especie, es decir cada lista aqui tiene 4 promedios
List_desv=[dessetosa,desversi,desvirgi] #Lista de desviaciones estándar de las 4 características por especie, es decir cada lista aqui tiene 4 datos

#Nota en general para casi todas mis listas de datos  
#0--longitud del sépalo
#1--ancho del sépalo
#2--longitud del pétalo
#3--ancho del pétalo



Especie=['Iris-setosa','Iris-versicolor','Iris-virginica']
Dat=['promedio de longitud del sépalo =','promedio de ancho del sépalo =','promedio de longitud del pétalo =','promedio de ancho del pétalo =']
DES=['Desvest longitud del sépalo =','Desvest de ancho del sépalo =','Desvest de longitud del pétalo =','Desvest de ancho del pétalo =']

#ciclo for para imprimir los valores promedio y desviaciones estándar según la especie
for l in range(len(Especie)):
    print('\n')
    print(Especie[l])
    for j in range(4):
        print(Dat[j],List_prom[l][j],'cm')
    print("\n")
    for i in range(4):
        print(DES[i],List_desv[l][i],'cm')

            
            
            
```

5) Ahora crea histogramas de densidad para cada característica (longitud del sépalo, ancho del sépalo, longitud del pétalo y ancho del pétalo), separados por especie de Iris. El código para crear un histograma de densidad se muestra a continuación:



```python

import matplotlib.pyplot as plt

#Listas vacías donde metemos sólo los datos de cada especie (cada una tiene 50)
SET=[] #50 datos para setosa
VER=[] #50 datos para versicolor
VIR=[] #50 datos para virginica

for j in range(4): #Repetimos por el número de columnas
    with open(r"C:\Users\DAVID\Desktop\iris\iris.data")  as f: #volvemos a abrir el archivo de texto
        for cols in csv.reader(f): #volvemos a leerlo por filas
            if cols[4]=="Iris-setosa": #condición de la primer flor
                    SET.append(float(cols[j]))#añadimos los elementos de setosa 
            elif cols[4]=='Iris-versicolor': #condición para la segunda flor
                    VER.append(float(cols[j])) #añadimos los elementos de versicolor
            elif cols[4]=='Iris-virginica': #condición para la tercera flor
                    VIR.append(float(cols[j])) #añadimos los elementos de virginica

                    
#Listas de datos de cada una de las 4 características según el tipo de flor                    
SET1=[] #longitud del sépalo para setosa
SET2=[] #Ancho del sépalo para setosa
SET3=[]#longitud del pétalo para setosa
SET4=[] #ancho del pétalo para setosa

VER1=[] #longitud del sépalo en versicolor
VER2=[] #Ancho del sépalo en versicolor
VER3=[] #longitud del pétalo en versicolor
VER4=[] #ancho del pétalo en versicolor

VIR1=[] # longitud del sépalo en virginica
VIR2=[] #Anchodel sépalo en virginica
VIR3=[] #longitud del pétalo en virginica
VIR4=[] #Ancho del pétalo en virginica

#ciclo for para  acomodar los datos de setosa en su respectiva característica
for i in range(len(SET)):
    if i>=0 and i<=49:
        SET1.append(SET[i])
    elif i>=50 and i<=99:
        SET2.append(SET[i])
    elif i>=100 and i<=149:
        SET3.append(SET[i])
    elif i>= 150 and i<=200:
        SET4.append(SET[i])
        
      
 #ciclo for para  acomodar los datos de versiclor en su respectiva característica

for i in range(len(VER)):
    if i>=0 and i<=49:
        VER1.append(VER[i])
    elif i>=50 and i<=99:
        VER2.append(VER[i])
    elif i>=100 and i<=149:
        VER3.append(VER[i])
    elif i>= 150 and i<=200:
        VER4.append(VER[i])
    
#ciclo for para  acomodar los datos de virginica en su respectiva característica                 
for i in range(len(VIR)):
    if i>=0 and i<=49:
        VIR1.append(VIR[i])
    elif i>=50 and i<=99:
        VIR2.append(VIR[i])
    elif i>=100 and i<=149:
        VIR3.append(VIR[i])
    elif i>= 150 and i<=200:
        VIR4.append(VIR[i])

#estas listas tienen los valores de longitud y ancho de los sépalos y pétalos juntando especies
S=[SET1,VER1,VIR1] #longitud del sépalo 
Ver=[SET2,VER2,VIR2] #Ancho del sépalo 
Vir=[SET3,VER3,VIR3] #longitud del pétalo 
BOa=[SET4,VER4,VIR4] #Ancho del pétalo

AAA=['Iris-setosa','Iris-versicolor','Iris-virginica'] #lista de nombres de las especies

fig,axes = plt.subplots(nrows=2, ncols=2,figsize=(10, 8)) #creamos una figura de 4 histogramas de tamaño 10x8

for i in range(len(S)): #como las listas S,Ver,Vir y BOa tienen el mismo tamaño, recorremos solo desde S
    axes[0, 0].hist(S[i], density=True,alpha=1,label=AAA[i]) #histograma de longitud del sépalo
    axes[0, 0].set_title('Longitud del sépalo')
    axes[0, 0].legend()

    axes[0, 1].hist(Ver[i], density=True,alpha=1,label=AAA[i])  #histograma del ancho del sépalo
    axes[0, 1].set_title('Ancho del sépalo')
    axes[0, 1].legend()

    axes[1, 0].hist(Vir[i], density=True,alpha=1,label=AAA[i]) #histograma de la longitud del pétalo
    axes[1, 0].set_title('Longitud del pétalo')
    axes[1, 0].legend()

    axes[1, 1].hist(BOa[i], density=True,alpha=1,label=AAA[i]) #histograma del ancho del pétalo 
    axes[1, 1].set_title('Ancho del pétalo')
    axes[1, 1].legend()
    
plt.tight_layout()
plt.show()






  
    
```

6 ) Una vez que tengas el histograma, puedes ajustar una curva de densidad
de probabilidad normal sobre  ́él. Para crear esta curva, primero necesitas
calcular la media (mean) y la desviación estándar (std) de los valores de
cada especie para la característica en cuestión.
Luego, genera una serie de valores x que cubran un rango alrededor de la
media:

```python

List_prom=[setosa,versicolor,virginica] #Lista con los valores promedio de las 4 características según la especie
List_desv=[dessetosa,desversi,desvirgi] #Lista de desviaciones estándar de las 4 características según la especie


X=[] #Lista donde guardaremos las 12 listas de puntos x para cada curva de cada histograma según la especie
for j in range(len(List_prom)):#buscamos entre los 3 elementos de la lista de promedios
    for i in range(len(setosa)): #buscamos entre los 4 elementos (promedios de cada característica), como setosa y las otras listas tienen la misma longitud, da igual si ponemos su longitud en el range
        x=np.linspace(List_prom[j][i]-5*List_desv[j][i],List_prom[j][i]+5*List_desv[j][i],100)
        X.append(x) #guardamos cada una de las listas de puntos generadas

#Dado que las listas que necesitamos para cada característica en cada histograma estan guardadas en X tanto en indices 
#pares como impares, y aprovechando que son solo 12 listas vamos a escribir las listas de puntos x que necesitaremos para cada histograma 
#manualmente
LS=[X[0],X[4],X[8]] #listas de puntos x para longitud del sépalo
AS=[X[1],X[5],X[9]] #Lista de puntos x para el ancho del sépalo
LP=[X[2],X[6],X[10]] #Lista de puntos x para la longitud del pétalo
AP=[X[3],X[7],X[11]] #Lista de puntos x para el ancho del pétalo


```

7) y 8)  Finalmente, puedes usar la función de densidad de probabilidad normal (norm.pdf()) para generar la curva y superponerla al histograma, además agregar un titulo y etiquetas a los ejes.

```python
from scipy.stats import norm #importamos la funcion de densidad de probabilidad


List_prom=[setosa,versicolor,virginica]
List_desv=[dessetosa,desversi,desvirgi]
Col=["blue",'orange','green'] #lista de colores para los histogramas y las curvas
fig,axes = plt.subplots(nrows=2, ncols=2,figsize=(10, 8)) #creamos las gráficas 
for i in range(len(S)): #nuevamente aprovechando que S, Ver, Vir y BOa tienen el mismo tamaño, iteramos sobre S
    axes[0, 0].hist(S[i], density=True,alpha=0.5,label=AAA[i],color=Col[i])
    axes[0, 0].plot(LS[i], norm(List_prom[i][0], List_desv[i][0]).pdf(LS[i]),color=Col[i]) #agregamos la curva
    axes[0, 0].set_title('Distribución de la longitud del sépalo')
    axes[0, 0].set_xlabel("Longitud (cm)")
    axes[0, 0].set_ylabel("Densidad")
    axes[0, 0].legend()
#Realizamos lo mismo para el resto de graficas,cambiando de lista de datos, promedio y desviación estándar
    axes[0, 1].hist(Ver[i], density=True,alpha=0.5,label=AAA[i],color=Col[i])
    axes[0, 1].set_title('Distribución del ancho del sépalo')
    axes[0, 1].plot(AS[i], norm(List_prom[i][1], List_desv[i][1]).pdf(AS[i]),color=Col[i])
    axes[0, 1].set_xlabel("Ancho (cm)")
    axes[0, 1].set_ylabel("Densidad")
    axes[0, 1].legend()

    axes[1, 0].hist(Vir[i], density=True,alpha=0.5,label=AAA[i],color=Col[i])
    axes[1, 0].set_title('Distribución de la longitud del pétalo')
    axes[1, 0].plot(LP[i], norm(List_prom[i][2], List_desv[i][2]).pdf(LP[i]),color=Col[i])
    axes[1, 0].set_xlabel("Longitud (cm)")
    axes[1, 0].set_ylabel("Densidad")
    axes[1, 0].legend()

    axes[1, 1].hist(BOa[i], density=True,alpha=0.5,label=AAA[i],color=Col[i])
    axes[1, 1].set_title('Distribución del Ancho del pétalo')
    axes[1, 1].plot(AP[i], norm(List_prom[i][3], List_desv[i][3]).pdf(AP[i]),color=Col[i])
    axes[1, 1].set_xlabel("Ancho (cm)")
    axes[1, 1].set_ylabel("Densidad")
    axes[1, 1].legend()

plt.tight_layout()
plt.show()

```
