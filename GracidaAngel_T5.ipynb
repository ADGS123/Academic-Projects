Primero vamos importar el conjunto de datos y realizar el chequeo habitual del conjunto de datos.

```python
import pandas as pd
from sklearn.datasets import load_wine #importamos los datos del vino
dataset = load_wine()
df = pd.DataFrame(data=dataset["data"], columns=dataset["feature_names"])
df["target"] = dataset["target"]
```

```python

#Calculamos los porcentajes de cada clase
class_counts = df['target'].value_counts().sort_index()
print("\nPorcentaje por clase:")
print((class_counts / class_counts.sum()) * 100)
```

```python
#Entrenamiento de los datos para arboles de desición
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

X = dataset["data"]
y = dataset["target"]

#División del conjunto de datos en prueba (20 %) y entrenamiento (80 %)
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, test_size=0.2, shuffle=False, random_state=42)
clf1 = DecisionTreeClassifier()
clf1.fit(X_train1, y_train1)
#imprimimos la métrica accuracy del arbol de desición
acc1 = accuracy_score(y_test1, clf1.predict(X_test1))
print("Accuracy previo:", acc1)
```

Entrena un modelo para cada una de las tres opciones y calcula la precisión. Compara y analiza estos valores con la precisión obtenida previamente.

```python
from sklearn.metrics import precision_score
#vemos las métricas precision_score y accuracy_score para los datos de entrenamiento
print("clases",[0,1,2])
#Caso sin mezcla
clf1 = DecisionTreeClassifier()
clf1.fit(X_train1, y_train1)
acc1 = precision_score(y_test1, clf1.predict(X_test1),average=None) #El average lo ponemos en None por que tenemos 3 clases mezcladas
print("precision sin mezcla:", acc1) #me va a devolver la métrica en un arreglo 1D


#Caso con mezcla
clf2 = DecisionTreeClassifier()
clf2.fit(X_train2, y_train2)
acc2 = precision_score(y_test2, clf2.predict(X_test2),average=None)
print("Precision con mezcla:", acc2)


#Caso con estratificación
clf3 = DecisionTreeClassifier()
clf3.fit(X_train3, y_train3)
acc3 = precision_score(y_test3, clf3.predict(X_test3),average=None)
print("precision con estrattificación:", acc3)

```

Como al final se me olvidó preguntar si se referia a accuracy en lugar de precision, pongo tambien la de accuracy

```python
from sklearn.metrics import accuracy_score
#Caso sin mezcla
clf1 = DecisionTreeClassifier()
clf1.fit(X_train1, y_train1)
acc1 = accuracy_score(y_test1, clf1.predict(X_test1))
print("accuracy sin mezcla:", acc1)

#Caso con mezcla
clf2 = DecisionTreeClassifier()
clf2.fit(X_train2, y_train2)
acc2 = accuracy_score(y_test2, clf2.predict(X_test2))
print("accuracy con mezcla:", acc2)


#Caso con estratificación
clf3 = DecisionTreeClassifier()
clf3.fit(X_train3, y_train3)
acc3 = accuracy_score(y_test3, clf3.predict(X_test3))
print("accuracy con estrattificación:", acc3)
```

Previamente, con accuracy obtuvimos un valor de 0.9166, que en realidad corresponde al caso sin mezclar, por otra parte, los valores de precisión aumentan con la mezcla y la estratificación representando una disminución en los errores de clasificación sobre el total de datos. Por otra parte, a pesar de que el caso sin mezcla solamente escogió datos de una misma clase en los datos de prueba, en el resto de casos (con mezcla y estratificación) la precision es mayor a 0.9, lo que indica que el modelo muy rara vez clasifica erróneamente los datos en sus respectivas clases.
Vemos que de hecho los porcentajes de datos tanto en el conjunto de entrenamiento como el de prueba del caso con estratificación esta mejor balanceado, por ello decimos que esta tercera opción proporciona un mejor rendimiento del método de clasificador por arbol de desición.

Estudia también las métricas precision, recall y f1-score.

```python
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report

#hacemos los reportes de clasificación, para los datls de prueba
print("Métricas del caso sin mezcla")
A1=classification_report(y_test1, clf1.predict(X_test1),output_dict=True) #agregamos el output_dic para que nos de un diccionario y quitemos las otras métricas que no nos interesan por ahora
A1.pop('accuracy', None) #borramos las que no ocupamos
A1.pop('macro avg', None)
A1.pop('weighted avg', None)
print(pd.DataFrame(A1).transpose()) #volvemos el diccionario a un dataframe para que no se vea descuadrado
print("\n") #repetimos lo mismo para los casos restantes
print("Métricas del caso con mezcla")
A2=classification_report(y_test2, clf2.predict(X_test2),output_dict=True)
A2.pop('accuracy', None)
A2.pop('macro avg', None)
A2.pop('weighted avg', None)
print(pd.DataFrame(A2).transpose())

print("\n")
print("Métricas del caso de estratificación")
A3=classification_report(y_test3, clf3.predict(X_test3),output_dict=True)
A3.pop('accuracy', None)
A3.pop('macro avg', None)
A3.pop('weighted avg', None)
print(pd.DataFrame(A3).transpose())
print("\n")
```

En general, el caso de estratificación ofrece un mejor balance de clases y un mejor rendimiento del clasificador, no obstante en los tres casos, la clase 2 fue la más sencilla de predecir. Todas las métricas tienen valores mayores a 0.9, en particular f1-score nos indica buen rendimiento, pensamos que dado que tenemos al rededor del 30 % de los datos de cada clase, el numero es menor, incrementando la posibilidad de no cometer errores. Con un conjunto de datos mas robusto seria una excelente manera de identificar la eficacia del modelo.

Cual es el valor más razonable? Por qué? ¿En qué casos es crítico utilizar
stratify=y y cuándo no es tan importante?

El valor mas razonable es el de estratificación, mantiene una proporcion muy cercana de datos en distintas clases, ademas de que necesita un shuffle=True, lo que indica que siempre mezclara los datos garantizando la no aparicion de un sesgo. Por lo tanto, decimos que es crítico usarlo en los casos donde tengamos que clasificar un conjunto de datos pero cuyos datos no representen una especie de ligadura o estructura fija, si no que podamos mezclarlos sin problemas para una mejor distribución en el entrenamiento y la prueba. Por el contrario no seria tan importante si los datos tuvieran algún tipo de patrón o estructura que no permitiese desordenarlos, como tiempos.
